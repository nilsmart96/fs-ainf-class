# Welcome to your third class - AI - The Frontier

## Goals

The course covers current topics in AI. Each year we will focus on 2-3 topics that are at the frontier of AI research and industrial application.

This year we have 3 Topics. The current content of the course is:

-	Deep Reinforcement Learning
-	Generative deep learning in terms of Generative Adversarial Neural Networks and diffusion models
- Graph Neural Netorks





## Limits
- Machine learning is a vast field and we can only cover some of the relevant topics



## After this class you should be able to..
- Be aware of important research streams in machine learning
- List main research challenges in machine learning
-	Identify and discuss current scientific and technical literature in deep learning.
-	Interpret research and debate the main findings of papers.
-	Chose an appropriate modelling structure for a novel problem based on the latest scientific literature.
-	Present research findings and their implications for a data-science project to others.

# Measurement of performance


## Class assignments (and their submission)
- For each block there is at least one assignment (continuous assessment)
- The continuous assessments have to be uploaded in canvas (a pdf of the relevant Jupyter Notebook AND the jupyter notebook itself), no later than end of day (Frankfurt time) 14 days after the respective class  
- Make sure the PDF does not have the text cut off on the sides or in the cells
- The assignments in this class are more challenging than the previous classes
- Start early with the continuous assignments, i.e. when we have covered the respective material in class
- You can help each other out with the continuous assessment, e.g. help a colleague to debug their code, but *you must not directly copy and paste code from others*. Such behaviour is plagiarism and will be awarded 0 marks  |
- Late submission: you get deducted 20% of the grade for every day that you are late. The rounding is up to the next day, so if you are one minute late it will be counted as one day. The ground truth for you submission time is the Canvas time stamp
